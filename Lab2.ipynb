```python
fileop = open("FEV.csv")
```

```python
labels = fileop.readline().strip().split(",")
```

```python
labels
```

```python
data = []
for l in fileop:
    data.append(l.strip().split(","))
```

```python
data
```

```python
l_id, l_age, l_fev, l_height, l_sex, l_smoke = [], [], [], [], [], []
for i in range(len(data)):
    l_id.append(int(data[i][0]))
    l_age.append(int(data[i][1]))
    l_fev.append(float(data[i][2]))
    l_height.append(float(data[i][3]))
    l_sex.append(data[i][4][1:-1])
    l_smoke.append(data[i][5])
```

```python
l_sex
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
male, female = 0, 0
for e in l_sex:
    if e == "female":
        female += 1
    else:
        male += 1
sex_values = [male, female]
sex_labels = 'Males', 'Females'
explode=(0, 0.02)
```

```python
plt.pie(sex_values, explode=explode, labels=sex_labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Percentage of Genders in Sample")
```

## Gender Relationships to other values

### Get the values of males/females and their FEV values

```python
male_FEV, female_FEV = [], []
for l in data:
    if l[4][1:-1] == "male":
        male_FEV.append(l[2])
    else:
        female_FEV.append(l[2])
```

```python
plt.plot(range(len(male_FEV)), male_FEV, label = "Male")
plt.plot(range(len(female_FEV)), female_FEV, color = 'r', label = "Female")
plt.xlabel("Individual Counts of People")
plt.ylabel("FEV")
plt.title("Graph of Gender to FEV")
plt.legend()
plt.tight_layout()
```

### This chart shows the relationship between individual counts of people to their FEV. We can see that males have a higher lung capacity then their female counterparts. This is very useful information if you are a male and wants to challenge a female to a breath holding competition!

### Now I plan to find the relationship between gender and heights.

```python
male_height, female_height = [], []
for l in data:
    if l[4][1:-1] == "male":
        male_height.append(float(l[3]))
    else:
        female_height.append(float(l[3]))
```

```python
plt.subplot(2,1,1)
plt.hist(male_height)
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.title("Male Height Relationship")
plt.subplot(2,1,2)
plt.hist(female_height, color = 'red')
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.title("Female Height Relationship")
plt.tight_layout(pad=0.4, w_pad=0.5,h_pad=1.0)
```

### This graph shows that females are generally between the height 60 to 65 whereas males can generally come in different sizes. However, we can see that there are about 1/3 of the male sample that are taller than all females. This is why finding a girlfriend is hard since every girl wants to find a man that is taller than herself.

### The next graph will probably show whether gender contribute to the habit of smoking.

```python
male_smoke, female_smoke = [], []
for l in data:
    if l[4][1:-1] == "male":
        male_smoke.append((l[5][1:-1]))
    else:
        female_smoke.append((l[5][1:-1]))
```

```python
male_smoke_yes, male_smoke_no = 0, 0
female_smoke_yes, female_smoke_no = 0, 0
for e in male_smoke:
    if e == "non-current smoker":
        male_smoke_no += 1
    else:
        male_smoke_yes += 1
for e in female_smoke:
    if e == "non-current smoker":
        female_smoke_no += 1
    else:
        female_smoke_yes += 1
```

```python
plt.subplot(1,2,1)
plt.pie([male_smoke_yes, male_smoke_no], explode=(0, 0.1), labels=["Smokers", "Non-smokers"], autopct='%1.1f%%', shadow=True, 
        startangle=90)
plt.title("Male Smokers and Non-Smokers")
plt.subplot(1,2,2)
plt.pie([female_smoke_yes, female_smoke_no], explode=(0, 0.1), labels=["Smokers", "Non-smokers"], autopct='%1.1f%%', shadow=True,
        startangle=90)
plt.title("Female Smokers and Non-Smokers")
plt.tight_layout()
```

### Surprisingly, we see that only 7.7% of the male sample smokes whereas 12.3% of the female sample smoke. Genderwise, it looks like females smoke more often then men. 

## Age Relationship to other values

```python
plt.bar(l_age,l_fev)
plt.xlabel("Age")
plt.ylabel("FEV")
plt.title("Age to FEV Relationship")
plt.tight_layout()
```

### This graph shows that the older we are, our FEV increases. This is quite logical since as we grow, our lungs also grow in capacity which result in a higher FEV.

### The next graph I plan to make is a graph showing the relationship of age to height. I believe that the graph will be similar to the graph above.

```python
plt.bar(l_age,l_height)
plt.xlabel("Age")
plt.ylabel("Height")
plt.title("Age to Height Relationship")
plt.tight_layout()
```

### The graph is indeed similar to the age vs FEV graph. We can see that maximum height is around the range of 70 to 75, and many people will reach that height at around their teens (15-18). We also know through the gender vs height graph that it is usually males who are growing to that particular range of height.

```python
l_smoke_digits = []
for e in l_smoke:
    if e[1:-1] == "current smoker":
        l_smoke_digits.append(1)
    else:
        l_smoke_digits.append(0)

plt.bar(l_age,l_smoke_digits)
plt.xlabel("Age")
plt.ylabel("0 for non-smoking, 1 for smoking")
plt.title("At what age do people smoke?")
plt.tight_layout()
```

### Surprisingly, apparently there are people who start to smoke at age 9! This sample shows that if you spot someone who smokes, they are usually age 9 or greater. There are quite alot of people who broke the rule of no smoking until 18. 

## Does your height affect your lung capacity?

```python
## Regression line help gotten from Tip Exploration Example
from scipy import stats
import numpy as np

slope, intercept, r_value, p_value, slope_std_error = stats.linregress(l_height,l_fev)
interps = np.linspace(40, 80, 200)


plt.scatter(l_height,l_fev)
plt.plot(interps, intercept + slope * interps, color = "red")
plt.xlabel("Height")
plt.ylabel("FEV")
plt.title("Height vs FEV")
plt.tight_layout()
```

### From the regression line, we can see that there is a strong relationship between Height and FEV. This can be extended to the graph of Age vs Height graph. As people increase in age, they will most oftenly grow in all aspects and their height is one of the aspects. Therefore an adult will have bigger lungs then a child which leads to this graph's findings. 

### Lastly, I plan to plot a graph to explore the relationship between smokers and their FEV

```python
fev_smoke, fev_nonsmoke = [], []
s = np.linspace(0, 6, 20)
for e in data:
    if e[5][1:-1] == "current smoker":
        fev_smoke.append(float(e[2]))
    else:
        fev_nonsmoke.append(float(e[2]))
```

```python
plt.hist(fev_nonsmoke,s, label = "Nonsmokers")
plt.hist(fev_smoke,s, label = "Smokers")
plt.xlabel("FEV")
plt.ylabel("Number of People")
plt.legend()
```

Only a small percentage of non smokers have a higher FEV then the top tier smokers. But that could be due to natural body composition. In fact, smokers are clustered in between 2 to 5 FEV whereas non smokers exists even below 2 FEV. However this result is probably caused by the relationship between age and smoking. As we have seen previously, the sample does not have anyone who smoke that is under the age of 9. And people who are under the age of 9 are the ones who have the smallest FEV due to their small lung capacity. Therefore, this graph doesnt really show anything concrete.
